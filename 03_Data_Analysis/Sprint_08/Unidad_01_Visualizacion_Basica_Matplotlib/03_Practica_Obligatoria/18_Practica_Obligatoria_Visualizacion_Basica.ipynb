#### #1.2  

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro.  

Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)  

2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand ITC", tamao 18, color azul acero o "steelblue")
3. El scatter de rea de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
# primero, vamos a agrupar las ciudades mas pobladas con groupby:
df_top_ciudades = (df_casas_california.groupby("city")["population_total"].sum().sort_values(ascending=False).head(5))
df_top_ciudades
```

```python
# barras con frecuencias absolutas de habitantes de cada ciudad 
plt.figure(figsize=(8, 8))
colores = ["#4C80D3", "#B7B7B7", "#F3A77C", "#CCE69E", "#DEB49A"]
bars = plt.bar(
    df_top_ciudades.index, # mis valores de barra x
    df_top_ciudades.values, # mis valores de barra y
    color=colores
    )

plt.xticks(rotation=0)
plt.ylabel("Habitantes (miles)")
plt.xlabel("Ciudades")
plt.title("Ciudades ms pobladas de California")

for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height,
        f"{int(height):,}".replace(",", "."),      # formato con separador de miles
        ha="center",
        va="bottom"
    )

plt.show()
```

```python
# fig.savefig("./data/barras_ciudades_mas_pobladas.png")
```

```python
# histograma de rea total por ciudad:
import matplotlib.font_manager as fm
fuente = fm.FontProperties(fname="/System/Library/Fonts/Supplemental/Bradley Hand Bold.ttf")

fig,axs = plt.subplots(1,1,figsize= (8,8))
sns.histplot(df_casas_california["area_land_sq_mi"],
             kde=False,
             color='steelblue',
             bins=50, 
             ax = axs,
             alpha= 0.7,
             edgecolor= None, # esto quita las lineas del histograma para que se vea continuo
            label= "Hist Area" # genera la leyenda usando el color de las barras
             ) 

axs.set_xlabel(r"Superficie (km$^{2}$)", fontsize=18, fontproperties= fuente)
axs.set_ylabel("nmero de ciudades", fontsize=18, fontproperties= fuente)
axs.legend(loc = "upper right")
plt.show()
```

```python
# fig.savefig("./data/histograma_area_total_ciudad.png")
```

```python
# scatter de rea de agua de ciudad por poblacin:

plt.scatter(df_casas_california["area_water_km2"],
            df_casas_california["population_total"]/1000, 
            c = "green", 
            alpha=0.5, 
            s= 1*df_casas_california["area_water_km2"])

plt.xlabel(r"Area de agua en la ciudad (km$^{2}$)")
plt.ylabel("Poblacin (miles habitantes)")
plt.xlim(0, 120)
plt.xticks(range(0, 121, 20))
plt.ylim(0, 4000)
plt.yticks(range(0, 4001, 500))
plt.title("Poblacin por kilmetro cuadrado de agua")

# en la clase ver cmo aadir la flecha porque no s calcular la ubicacin de los outliers

plt.show();
```

```python
# fig.savefig("./data/scatter_poblacion_por_km2.png")
```

```python
#boxplot de altura de las ciudades:

plt.boxplot(df_casas_california["elevation_m"].dropna(), # dropna para quitar los null que hay
            whis=1.5,
            vert=False) # para que se dibuje horizontalmente
plt.xlabel("Altitud")
plt.xlim(0, 2500)
plt.title("Diagrama de caja de la altitud de las ciudades de California") 
plt.show();

# no s cmo pintarlo de rojo, lo ver en clase
```
