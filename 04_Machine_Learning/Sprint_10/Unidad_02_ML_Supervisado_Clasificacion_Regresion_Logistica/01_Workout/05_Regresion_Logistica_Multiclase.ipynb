## Clasificacin Multicategrica con Regresin Logstica

### Entender el problema

Para mostrar la creacin de un modelo multiclase y su evaluacin vamos a tratar un problema en el que queremos detectar cul es el sistema operativo de los usuarios que acceden a una direccin web especfica (y que no es importante para el problema) en la que pasan un tiempo determinado, navegan a travs de las pginas del site, realizan un conjunto de acciones y dejan una valoracin de la misma. Encontrars una variable "clase" que contiene el target, esta vez no tendrs que buscarlo, y que toma tres posibles valores en funcin del OS: Windows, Linux, Mac. 

Creemos ese modelo predictor a partir de una regresin logstica.

### Split

Son muy pocos datos (170) pero vamos a hacer el split igual para que vayamos cogiendo el hbito. Adems estratificamos sobre el target (pero ojo esto es para hacer ms limpio el ejercicio, por ejemplo porque suponemos que es algo que se va a mantener debido al marketshare de cada sistema operativo)

```python
train_set, test_set = train_test_split(df_os, test_size= 0.2, stratify= df_os["clase"], random_state= 42)
```

```python
train_set["clase"].value_counts(normalize = True)
```

```python
test_set["clase"].value_counts(normalize = True)
```

### Mini-EDA

Vamos a ver la correlacin de las variables con el target, pero realmente son tan pocas que no vamos a quitar ninguna:

```python
train_set.drop("clase", axis = 1).hist()
plt.show()
```

```python
sns.pairplot(train_set,
            hue='clase',
            height=4,
            vars=["duracion", "paginas","acciones","valor"],
            kind='reg'); # "reg" de regresin lineal
```

```python
features = train_set.drop("clase", axis = 1).columns.to_list() #lista de features, sin la columna objetivo, y las aado a una lista
print(features)
```

### Procesamiento de Features

```python
for col in features: # aplicar logaritmo natural
    train_set[col] = train_set[col].apply(np.log) #logaritmo natural
train_set[features].hist() #ver histograma tras transformacin
```

```python
for col in features:
    test_set[col] = test_set[col].apply(np.log)
# Ojo ni lo mires :-), pero tienes que aplicarle las mismas transformaciones que el train.
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(train_set[features]) # Adaptamos el StandarScaler al Train, pero luego usamos el mismo en el test (porque ya la media y la desviacin son las del train)
X_train = pd.DataFrame(scaler.transform(train_set[features]), columns = features)
X_test = pd.DataFrame(scaler.transform(test_set[features]), columns= features)
X_train.describe()
```

```python
y_train = train_set["clase"] # separo variable objetivo
y_test = test_set["clase"] #separo variable objetivo
```

### Creamos el modelo

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(max_iter=10000) 

clf.fit(X_train, y_train) # No hay que decir nada en especial, el detecta el tipo de clasificador multicategrico y aplica el one-vs-rest o one-vs-all
```

Si mostramos los coeficientes o pesos del modelo veremos que en realidad ha entrenado un modelo de regresin logstica para cada clase como vimos en la sesin de teora. Ha hecho un one-vs-rest

```python
clf.coef_ #coeficientes de cada variable para cada clase, lo que hace es ajustar una frontera de decisin para cada clase
# entreno tres modelos de clasificacin binaria (one-vs-rest), en el primero la clase 0 contra las clases 1 y 2, 
# en el segundo la clase 1 contra las clases 0 y 2, y en el tercero la clase 2 contra las clases 0 y 1.
#y luego la clase que me d la probabilidad ms alta es la que asigno.
```

```python
#para ver el impacto de cada variable en cada clase

df_coefs = pd.DataFrame(clf.coef_, columns=X_train.columns)

fig, ax = plt.subplots(1, 3, figsize=(10, 3), sharex=True)
for i, subdf in df_coefs.iterrows():
    subdf.plot(kind="barh", ax=ax[i])
    ax[i].set_title(f"Parmetros clase {i}")

fig.tight_layout()
```

Ten en cuenta que estos coeficientes estn aplicados sobre el logaritmo de las variables, an as puedes ver que influye en la seleccin de cada clase.

Podemos ver las predicciones

```python
clf.predict(X_train) #predice una clase para cada fila del X_train
```

Y las probabilidades para cada instancia, nos da un vector de tres probabilidades una para cada clase y puedes comprobar que escoge la de mayor probabilidad:

```python
clf.predict_proba(X_train) #para cada fila me da la probabilidad de cada clase
#y escoge la clase con mayor probabilidad
```

```python
from sklearn.metrics import accuracy_score

acc_train = accuracy_score(y_train, clf.predict(X_train))
print(acc_train)
```

En este caso vemos que el accuracy es superior a la contribucin de la clase mayoritaria, no tiene mala pinta (pero recuerda que es el train). No vamos a ver el test, antes vamos a usar la validacin cruzada y luego veremos las mtricas para multiclase.

***

### Validamos el modelo

Usemos la validacin cruzada no tanto para comparar con otros modelos, que no tenemos, sino como para adelantarnos a la evaluacin con el test y de nuevo ir cogiendo hbito.

```python
from sklearn import model_selection
name='Logistic Regression'
cv_results = model_selection.cross_val_score(clf, X_train, y_train, cv=5, scoring='accuracy')

msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
print(cv_results)
print(msg)
#nos sale una accuracy de 67% con validacin cruzada, lo cual no est nada mal
```

Ya vemos que sale menor que el 72% del train. 

### Evaluacin del modelo

```python
from sklearn.metrics import precision_score, recall_score, accuracy_score, balanced_accuracy_score,\
                            roc_auc_score, roc_curve, precision_recall_curve, confusion_matrix, ConfusionMatrixDisplay
#creamos la prediccin sobre el conjunto de test
```

```python
# Prediccin sobre el conjunto de test
y_proba = clf.predict_proba(X_test)
y_pred = clf.predict(X_test)
```

#### Matriz de confusion

A mano, como un dataframe a partir de los datos que nos da el `confusion_matrix` de sklearn:

```python
# Matriz de confusin, para ver cmo de bien hemos clasificado cada clase
cm = pd.DataFrame(confusion_matrix(y_test, y_pred), index=[f"Real {i}" for i in clf.classes_], 
                  columns=[f"Predicho {i}" for i in clf.classes_])

cm
# en el predicho 0, la precisin es 14/17, una precisin del 82%
# en el predicho 1, la precisin lo ha clavado, es 5/5, una precisin del 100%
# en el predicho 2, la precisin es 9/12, una precisin del 75%
#siendo un test pequeo, est bastante bien pero no es concluyente
```

Grficamente:

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred) #vemos grficamente la tabla de matriz de confusin
```

```python
# y podemos sacarla en porcentajes: nos da la precisin por clase
sns.heatmap(cm/np.sum(cm), annot=True, 
            fmt='.2%', cmap='Blues');
```

### Mtricas MACRO y MICRO

```python
#macro calcula los valores por clase y luego hace la media aritmtica. Para cuando tenemos clases desbalanceadas
# la micro calcula globalmente los TP, TN, FP, FN y luego calcula la mtrica. Para cuando las clases estn balanceadas

micro_prec = precision_score(y_test, y_pred, average="micro") # average puede ser 'micro', 'macro', 'weighted', 'samples'
print("MICRO PRECISION:", micro_prec)

macro_prec = precision_score(y_test, y_pred, average="macro")
print("MACRO PRECISION:", macro_prec)

micro_rec = recall_score(y_test, y_pred, average="micro")
print("MICRO RECALL:", micro_rec)

macro_rec = recall_score(y_test, y_pred, average="macro")
print("MACRO RECALL:", macro_rec)

micro_acc = accuracy_score(y_test, y_pred)
print("MICRO ACCURACY:", micro_acc)
```

```python
# aqui vemos la accuracy final
acierto = accuracy_score(y_test, y_pred)

error = 1 - acierto
print("Acierto:", round(acierto*100, 2), "%")
print("Error:", round(error*100, 2), "%")
# es muy alta porque el dataset es pequeo y sencillo
```
